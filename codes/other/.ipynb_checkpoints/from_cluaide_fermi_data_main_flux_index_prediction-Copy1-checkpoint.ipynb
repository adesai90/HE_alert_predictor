# Starting prediction!

```python
# Make new feature files or sue old ones:
make_feature_files=True

# Set Debug Flag
debug_this=False


seed=123
if os.path.isdir(f'seed_{seed}/')!=True:
    os.mkdir(f'seed_{seed}/')
```

```python
for src_index in range(len(src_names)):
    if make_feature_files==False:
        continue
    if src_names[src_index]!="4FGL_J1321.1+2216":#  "4FGL_J0118.3-6008":
        continue
    
    alerts_data = np.load(alerts_per_bin_files[src_index])
    all_times = alerts_data[0]  # MJD times
    all_flux = alerts_data[1]   # Flux values
    all_index = alerts_data[2]  # Index values
    
    # Create DataFrame with flux and index
    df = pd.DataFrame({
        '#MJD': all_times,
        'flux': all_flux,
        'index': all_index
    })

    df.set_index('#MJD', inplace=True)
    print(all_times, all_flux)
    
    print(f"\n{'='*70}")
    print(f"Processing: {src_names[src_index]}")
    print(f"{'='*70}")
    print(f"Total data points: {len(df)}")
    print(f"Non-zero flux points: {(df['flux'] > 0).sum()} ({100*(df['flux'] > 0).sum()/len(df):.1f}%)")
    print(f"Non-zero index points: {(df['index'] != 0).sum()} ({100*(df['index'] != 0).sum()/len(df):.1f}%)")
    
    if debug_this==True:
        df.plot(figsize=(10,5))

    # ---- CREATE FEATURES - USE PERCENTAGE CHANGE FOR FLUX ----
    batch_df = pd.DataFrame(index=df.index)
    
    # Calculate percentage change for flux (better signal than absolute change)
    df['flux_change'] = df['flux'].pct_change(1).fillna(0)
    df['flux_change'] = df['flux_change'].replace([np.inf, -np.inf], 0)
    
    # For index, absolute change is fine
    df['index_change'] = df['index'].diff(1).fillna(0)
    
    # CURRENT values (no delay!)
    batch_df['flux_current'] = df['flux']
    batch_df['index_current'] = df['index']
    
    # Only lag_1 for minimal delay
    batch_df['flux_lag_1'] = df['flux'].shift(1)
    batch_df['index_lag_1'] = df['index'].shift(1)
    
    # CHANGE features (percentage for flux, absolute for index)
    batch_df['flux_change_current'] = df['flux_change']  # NO LAG!
    batch_df['flux_change_lag_1'] = df['flux_change'].shift(1)
    batch_df['index_change_lag_1'] = df['index_change'].shift(1)
    
    # Momentum (rate of change of change)
    batch_df['flux_momentum'] = df['flux_change'].diff(1).fillna(0)
    batch_df['index_momentum'] = df['index_change'].diff(1).fillna(0)
    
    # Simple moving average
    batch_df['flux_ma_3'] = df['flux'].rolling(window=3, min_periods=1).mean()
    batch_df['index_ma_3'] = df['index'].rolling(window=3, min_periods=1).mean()
    
    # Distance from moving average (as percentage for flux)
    batch_df['flux_dist_from_ma'] = ((df['flux'] - batch_df['flux_ma_3']) / (batch_df['flux_ma_3'] + 1e-10)).fillna(0)
    batch_df['index_dist_from_ma'] = (df['index'] - batch_df['index_ma_3']).fillna(0)
    
    # TARGETS - percentage change for flux, absolute change for index
    batch_df['flux_change_target'] = df['flux'].pct_change(1).shift(-1).fillna(0).replace([np.inf, -np.inf], 0)
    batch_df['index_change_target'] = df['index'].diff(1).shift(-1).fillna(0)
    
    # Keep absolute targets for reconstruction and metrics
    batch_df['flux_target'] = df['flux'].shift(-1)
    batch_df['index_target'] = df['index'].shift(-1)
    
    # Keep CURRENT values for reconstruction
    batch_df['flux_current_for_recon'] = df['flux']
    batch_df['index_current_for_recon'] = df['index']
    
    # Drop NaN rows
    batch_df = batch_df.dropna()
    
    if len(batch_df) < 5:
        print(f"Warning: {src_names[src_index]} has insufficient data")
        #continue
    
    print(f"Samples after feature engineering: {len(batch_df)}")
    print(f"\nFlux CHANGE (percentage) target statistics:")
    print(f"  Min: {batch_df['flux_change_target'].min():.4f}")
    print(f"  Max: {batch_df['flux_change_target'].max():.4f}")
    print(f"  Mean: {batch_df['flux_change_target'].mean():.4f}")
    print(f"  Std: {batch_df['flux_change_target'].std():.4f}")
    
    print(f"\nIndex CHANGE (absolute) target statistics:")
    print(f"  Min: {batch_df['index_change_target'].min():.3f}")
    print(f"  Max: {batch_df['index_change_target'].max():.3f}")
    print(f"  Mean: {batch_df['index_change_target'].mean():.3f}")
    print(f"  Std: {batch_df['index_change_target'].std():.3f}")
    
    # Create feature and target dataframes
    directory = f'seed_{seed}/feature_store'
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    id_val = "v2"
    
    # Features for FLUX prediction
    flux_features_cols = [
        'flux_current', 'flux_lag_1',
        'flux_change_current', 'flux_change_lag_1',
        'flux_momentum', 'flux_ma_3', 'flux_dist_from_ma',
        'index_current', 'index_lag_1',
        'index_change_lag_1', 'index_momentum'
    ]
    flux_features_df = batch_df[flux_features_cols].copy()
    flux_features_df.to_csv(f'{directory}/{src_names[src_index]}_flux_features_{id_val}.csv')
    
    # Targets for FLUX prediction
    flux_targets_df = batch_df[['flux_change_target', 'flux_target', 'flux_current_for_recon']].copy()
    flux_targets_df.to_csv(f'{directory}/{src_names[src_index]}_flux_targets_{id_val}.csv')
    
    # Features for INDEX prediction
    index_features_cols = [
        'index_current', 'index_lag_1',
        'index_change_lag_1', 'index_momentum',
        'index_ma_3', 'index_dist_from_ma',
        'flux_current', 'flux_lag_1',
        'flux_change_lag_1', 'flux_momentum'
    ]
    index_features_df = batch_df[index_features_cols].copy()
    index_features_df.to_csv(f'{directory}/{src_names[src_index]}_index_features_{id_val}.csv')
    
    # Targets for INDEX prediction
    index_targets_df = batch_df[['index_change_target', 'index_target', 'index_current_for_recon']].copy()
    index_targets_df.to_csv(f'{directory}/{src_names[src_index]}_index_targets_{id_val}.csv')
    
    print(f"Created feature files for {src_names[src_index]} - {len(batch_df)} samples")
```

```python
def train_model(fermi_src, directory, id_val, model_type='flux'):
    """
    Train XGBoost model - flux uses percentage change, index uses absolute change
    """
    if fermi_src != "4FGL_J1321.1+2216":#  "4FGL_J0118.3-6008":
        return
    
    feature_file = f'{directory}/{fermi_src}_{model_type}_features_{id_val}.csv'
    target_file = f'{directory}/{fermi_src}_{model_type}_targets_{id_val}.csv'
    
    X_train = pd.read_csv(feature_file)
    X_train.set_index('#MJD', inplace=True)
    
    targets_df = pd.read_csv(target_file)
    targets_df.set_index('#MJD', inplace=True)
    
    # Use all data except last points for training
    X_train_mod = X_train.head(leave_number_indices)
    targets_df_mod = targets_df.head(leave_number_indices)
    
    # Train on CHANGE target (percentage for flux, absolute for index)
    y_train = targets_df_mod[f'{model_type}_change_target']
    
    # Remove any remaining NaN values
    mask = ~(X_train_mod.isna().any(axis=1) | y_train.isna())
    X_train_mod = X_train_mod[mask]
    y_train = y_train[mask]
    
    if len(X_train_mod) < 5:
        print(f"Warning: {fermi_src} has insufficient data for {model_type} training")
        return
    
    print(f"\n{'='*70}")
    print(f"Training {model_type.upper()} model for {fermi_src}")
    print(f"  Target: {'Percentage Change' if model_type == 'flux' else 'Absolute Change'}")
    print(f"{'='*70}")
    print(f"Training samples: {len(X_train_mod)}")
    print(f"Target statistics:")
    print(f"  Min:  {y_train.min():.6f}")
    print(f"  Max:  {y_train.max():.6f}")
    print(f"  Mean: {y_train.mean():.6f}")
    print(f"  Std:  {y_train.std():.6f}")
    
    # Model parameters
    params = {
        'objective': 'reg:squarederror',
        'eval_metric': 'rmse',
        'eta': 0.3,
        'max_depth': 3,
        'min_child_weight': 1,
        'subsample': 0.9,
        'colsample_bytree': 0.9,
        'gamma': 0,
        'reg_alpha': 0.05,
        'reg_lambda': 0.5,
        'tree_method': 'hist',
    }
    
    dtrain = xgb.DMatrix(data=X_train_mod, label=y_train)
    
    # Cross-validation
    cv_results = xgb.cv(
        params=params,
        dtrain=dtrain,
        num_boost_round=200,
        nfold=3,
        early_stopping_rounds=30,
        metrics='rmse',
        as_pandas=True,
        seed=seed,
        verbose_eval=False
    )
    
    optimal_rounds = max(cv_results['test-rmse-mean'].idxmin() + 1, 20)
    best_rmse = cv_results['test-rmse-mean'].min()
    
    print(f"\nCross-validation results:")
    print(f"  Best RMSE: {best_rmse:.6f}")
    print(f"  Optimal boost rounds: {optimal_rounds}")
    
    # Train final model
    final_model = xgb.XGBRegressor(
        n_estimators=optimal_rounds,
        **params,
        random_state=seed
    )
    
    final_model.fit(X_train_mod, y_train)
    
    # Check predictions
    train_pred = final_model.predict(X_train_mod)
    pred_std = np.std(train_pred)
    pred_range = train_pred.max() - train_pred.min()
    pred_mean = np.mean(train_pred)
    
    print(f"\nTraining predictions:")
    print(f"  Min:   {train_pred.min():.6f}")
    print(f"  Max:   {train_pred.max():.6f}")
    print(f"  Mean:  {pred_mean:.6f}")
    print(f"  Std:   {pred_std:.6f}")
    print(f"  Range: {pred_range:.6f}")
    
    if pred_std < 1e-6:
        print("  ⚠️  WARNING: Predictions have very low variance")
    else:
        print("  ✓ Predictions show variance")
    
    # Check feature importance
    importance = final_model.feature_importances_
    feature_names = X_train_mod.columns
    print(f"\nTop 5 important features:")
    for i in np.argsort(importance)[-5:][::-1]:
        print(f"  {feature_names[i]}: {importance[i]:.4f}")
    
    # Save model
    if not os.path.isdir(f'seed_{seed}/models'):
        os.mkdir(f'seed_{seed}/models')
    
    filename = f'seed_{seed}/models/{fermi_src}_{model_type}_model_{id_val}.pkl'
    joblib.dump(final_model, filename)
    
    print(f"\n✓ Saved model: {filename}")
    print(f"{'='*70}\n")
    
    del dtrain, cv_results
```

```python
# Train models
make_training_files = True
id_val = "v2"
directory = f'seed_{seed}/feature_store'

if make_training_files:
    for src_index in range(len(src_names)):
        if src_names[src_index] != "4FGL_J1321.1+2216":#  "4FGL_J0118.3-6008":
            continue
        
        print(f"\n{'='*60}")
        print(f"Training models for {src_names[src_index]}")
        print(f"{'='*60}")
        
        train_model(src_names[src_index], directory, id_val, model_type='flux')
        train_model(src_names[src_index], directory, id_val, model_type='index')
```

```python
def make_predictions_plot(src_names, directory, id_val, model_type='flux'):
    """
    Make predictions and reconstruct absolute values
    """
    
    for src_index in range(len(src_names)):
        src = src_names[src_index]
        if src != "4FGL_J1321.1+2216":#  "4FGL_J0118.3-6008":
            continue
        
        try:
            print(f"\n{'='*70}")
            print(f"Making {model_type.upper()} predictions for {src}")
            print(f"{'='*70}")
            
            # Load features
            features_df = pd.read_csv(f'{directory}/{src}_{model_type}_features_{id_val}.csv')
            features_df.set_index('#MJD', inplace=True)
            
            # Load model and predict CHANGE
            model_file = f'seed_{seed}/models/{src}_{model_type}_model_{id_val}.pkl'
            model = joblib.load(model_file)
            predicted_change = model.predict(features_df)
            
            print(f"\nPredicted CHANGE statistics:")
            print(f"  Min:   {predicted_change.min():.6f}")
            print(f"  Max:   {predicted_change.max():.6f}")
            print(f"  Mean:  {predicted_change.mean():.6f}")
            print(f"  Std:   {predicted_change.std():.6f}")
            
            
            # Load targets
            targets_df = pd.read_csv(f'{directory}/{src}_{model_type}_targets_{id_val}.csv')
            targets_df.set_index('#MJD', inplace=True)
            
            # RECONSTRUCT absolute predictions
            # For flux: predicted = current * (1 + pct_change)
            # For index: predicted = current + change
            current_values = targets_df[f'{model_type}_current_for_recon'].values
            
            if model_type == 'flux':
                predicted_absolute = current_values * (1 + predicted_change)
            else:
                predicted_absolute = current_values + predicted_change
            
            print(f"\nReconstructed ABSOLUTE predictions:")
            print(f"  Min:   {predicted_absolute.min():.6e}")
            print(f"  Max:   {predicted_absolute.max():.6e}")
            print(f"  Mean:  {predicted_absolute.mean():.6e}")
            print(f"  Std:   {predicted_absolute.std():.6e}")
            
            # Create results dataframe
            predictions_df = pd.DataFrame({
                f'{model_type}_change_predicted': predicted_change,
                f'{model_type}_predicted': predicted_absolute
            }, index=features_df.index)
            
            result_df = predictions_df.join(targets_df, how='inner')
            print(result_df)
            
            # Split train/test
            train_df = result_df.head(leave_number_indices)
            test_df = result_df.tail(-leave_number_indices)
            
            print(f"\nTest set size: {len(test_df)}")
            if len(test_df) > 0:
                test_preds = test_df[f'{model_type}_predicted'].values
                print(f"  Predicted Min:  {test_preds.min():.6e}")
                print(f"  Predicted Max:  {test_preds.max():.6e}")
                print(f"  Predicted Mean: {test_preds.mean():.6e}")
            
            # Create plot
            fig = plt.figure(figsize=(14, 8), dpi=150)
            ax = plt.subplot()
            
            # Plot training data
            ax.plot(train_df.index.values, train_df[f'{model_type}_target'].values, "-o", 
                    color="blue", linewidth=2, markersize=4, label="Training Data", alpha=0.7)
            ax.plot(train_df.index.values, train_df[f'{model_type}_predicted'].values, "--s", 
                    color="cyan", linewidth=2, markersize=4, label="Training Predictions", alpha=0.7)
            
            # Plot test data
            if len(test_df) > 0:
                ax.plot(test_df.index.values, test_df[f'{model_type}_target'].values, "-o", 
                        color="black", linewidth=2, markersize=6, label="Test Data", alpha=0.9)
                ax.plot(test_df.index.values, test_df[f'{model_type}_predicted'].values, "--s", 
                        color="red", linewidth=2, markersize=6, label="Test Predictions", alpha=0.9)
            
            plt.xlabel("Time (MJD)", fontsize=16, fontweight='bold')
            plt.ylabel(f"{model_type.capitalize()} Value", fontsize=16, fontweight='bold')
            title_suffix = "% Change" if model_type == 'flux' else "Change"
            plt.title(f"{src} - {model_type.upper()} Prediction (via {title_suffix})", 
                     fontsize=18, fontweight='bold')
            plt.legend(fontsize=12, loc='best')
            plt.grid(True, alpha=0.3)
            plt.tight_layout()
            
            plt.savefig(f'seed_{seed}/predictions/{src}_{model_type}_prediction_{id_val}.pdf', 
                       bbox_inches="tight", dpi=150)
            plt.show()
            
            # Calculate metrics on test set
            if len(test_df) > 0:
                from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
                
                # Metrics on CHANGE
                mse_change = mean_squared_error(test_df[f'{model_type}_change_target'], 
                                                test_df[f'{model_type}_change_predicted'])
                mae_change = mean_absolute_error(test_df[f'{model_type}_change_target'], 
                                                 test_df[f'{model_type}_change_predicted'])
                
                # Metrics on ABSOLUTE (reconstructed)
                mse_absolute = mean_squared_error(test_df[f'{model_type}_target'], 
                                                  test_df[f'{model_type}_predicted'])
                mae_absolute = mean_absolute_error(test_df[f'{model_type}_target'], 
                                                   test_df[f'{model_type}_predicted'])
                r2_absolute = r2_score(test_df[f'{model_type}_target'], 
                                      test_df[f'{model_type}_predicted'])
                
                print(f"\nTest metrics (CHANGE):")
                print(f"  MSE: {mse_change:.6f}")
                print(f"  MAE: {mae_change:.6f}")
                
                print(f"\nTest metrics (ABSOLUTE reconstructed):")
                print(f"  MSE: {mse_absolute:.6e}")
                print(f"  MAE: {mae_absolute:.6e}")
                print(f"  R²:  {r2_absolute:.4f}")
            
            print(f"\n✓ Created prediction plot for {src} ({model_type})")
            print(f"{'='*70}\n")
            
        except Exception as e:
            print(f"ERROR processing {src} ({model_type}): {str(e)}")
            import traceback
            traceback.print_exc()
            continue

```

```python
# Make predictions for both flux and index
make_predictions_plot(src_names, directory, id_val, model_type='flux')
make_predictions_plot(src_names, directory, id_val, model_type='index')

print("\nAll predictions completed!")
```
